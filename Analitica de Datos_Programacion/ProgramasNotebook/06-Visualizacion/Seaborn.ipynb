```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as pit
```

```python
df = pd.read_csv('./datasets/melbourne_housing-clean.csv',index_col=0)
```

```python
df = pd.DataFrame(
    {"Price 1": [7, 1, 5, 6, 3, 10, 5, 8], "Day": [1, 2, 3, 4, 5, 6, 7, 8]}
)

s1 = sns.lineplot(x="Day", y="Price 1", data=df, color="red")
```

```python
df = pd.DataFrame(
    {
        "Precio": [7, 1, 5, 6, 3, 10, 5, 8],
        "Canciones": ["C1", "C2", "C1", "C2", "C1", "C2", "C1", "C2"],
        "Artist": [1, 1, 2, 2, 3, 3, 4, 4],
    }
)

s = sns.lineplot(x="Artist", y="Precio", data=df, hue="Canciones")
```

```python
df = pd.DataFrame(
    {
        "Price 1": [7, 1, 5, 6, 3, 10, 5, 8],
        "Price 2": [1, 2, 8, 4, 3, 9, 5, 2],
        "Day": [1, 2, 3, 4, 5, 6, 7, 8],
    }
)

s1 = sns.lineplot(x="Day", y="Price 1", data=df, color="red")

s2 = sns.lineplot(x="Day", y="Price 2", data=df, color="blue")

pit.legend(labels=["Price1", "Price2"])
```

# Count plot vertical

```python
data = ["A", "A", "B",
        "B", "B", "C"]

# Count plot
sns.countplot(x = data);
```

```python
data = ["A", "A", "B",
        "B", "B", "C"]

# Count plot
sns.countplot(y = data);
```

# Count plot por grupo y subgrupo

```python
data = ["A", "A", "B", "A",
        "B", "B", "C", "C"]
group = ["G1", "G1", "G2", "G2",
         "G1", "G2", "G1", "G2"]

# Count plot
sns.countplot(x = data, hue = group)
```

# Personalizacin de color

```python
data = ["A", "A", "B",
        "B", "B", "C"]

# Count plot
sns.countplot(x = data, saturation = 5, color = 'red');
```

# Paleta de colores

```python
data = ["A", "A", "B",
        "B", "B", "C"]

# Count plot
sns.countplot(x = data, palette=  "Set1")
```

# Un unico color para todas las barras

```python
data = ["A", "A", "B",
        "B", "B", "C"]

# Count plot
sns.countplot(x = data,
              color = '#E333FF')
```

# Color de borde

```python
data = ["A", "A", "B",
        "B", "B", "C"]

# Count plot
sns.countplot(x = data,
              edgecolor = "yellow");
```
