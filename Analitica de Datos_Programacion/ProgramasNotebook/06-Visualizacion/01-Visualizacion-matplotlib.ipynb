## Matplotlib's two interfaces


### Imperative interface


```python
x = np.linspace(0,2,100)
plt.figure(figsize=(8,6))
plt.plot(x,x, label ='Linear', color = '#c0261c', linewidth = 1.5, linestyle = '-.')
plt.plot(x, np.cos(x), label = 'cosine',color= '#c59538', linewidth = 2 , linestyle = ':')
plt.plot(x, np.sin(x), label ='sine',color= '#c0d860' )
plt.xlabel('x Label', size = 14)
plt.ylabel('y Label', size = 14)
plt.title('Graficas de Lineas', size = 20)
plt.legend(loc= 'lower left')
```
